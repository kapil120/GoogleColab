```
import pandas as pd
```

```
data1=pd.read_csv("mobile_cleaned.csv")
data1.head()
```

```
ax=sns.scatterplot(x="stand_by_time",y="battery_capacity",data=data1)
```

```
ax=sns.scatterplot(x="stand_by_time",y="battery_capacity",hue="thickness",data=data1)
```

```
sx=sns.distplot(data1["stand_by_time"]) #In sns distplot is use for histogram
```

```
#kde =it will remove the distribution 
#rug =it will show where the exact values are or it will show the exect values
sx=sns.distplot(data1["stand_by_time"],kde=False,rug=True)
```

```
jx =sns.boxplot(x="is_liked",y="battery_capacity",data=data1)
```

```
sx =sns.boxplot(x="expandable_memory",y="price",data=data1)
```
